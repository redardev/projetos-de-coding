```python
def soma(n1,n2):
    return n1+n2
def subtrai(n1,n2):
    return n1-n2
def divide(n1,n2):
    return n1/n2
def multiplica(n1,n2):
    return n1*n2
def potencia(n1, n2):
    return n1 ** n2
def raiz_quadrada(n):
    return n ** 0.5
def modulo(n1, n2):
    return abs(n1 - n2)

    
```

```python
while True:
    print('+==================+')
    print('| MENU DE OPERAÇÕES')
    print('|  1 - somar       ')
    print('|  2 - subtrair    ')
    print('|  3 - multiplicar ')
    print('|  4 - divisão     ')
    print('|  5 - potência    ')
    print('|  6 - raiz quadrada')
    print('|  7 - módulo      ')
    print('|  0 - sair        ')
    print('+==================+')

    op = int(input())
    if op == 0:
        break
    if op in [1, 2, 3, 4, 5, 7]:
        n1 = int(input('Primeiro Número: '))
        n2 = int(input('Segundo Número: '))
    elif op in [6]:
        n = int(input('Número: '))

    if op == 1:
        print('O resultado da soma é: ', soma(n1, n2))
    elif op == 2:
        print('O resultado da subtração é: ', subtrai(n1, n2))
    elif op == 3:
        print('O resultado da multiplicação é: ', multiplica(n1, n2))
    elif op == 4:
        print('O resultado da divisão é: ', divide(n1, n2))
    elif op == 5:
        print('O resultado da potência é: ', potencia(n1, n2))
    elif op == 6:
        print('O resultado da raiz quadrada é: ', raiz_quadrada(n))
    elif op == 7:
        print('O resultado do módulo é: ', modulo(n1, n2))
    else:
        print('Esta opcao e invalida, tente outra')
        
```
